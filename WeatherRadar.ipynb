```python
import pandas as pd
import united_states
```

```python
w2 =  pd.read_csv('data/weather_radar_stations.csv')
# Change names to something that can just be readily used by the module.
w2.rename(columns={'y': 'lat','x':'lon'}, inplace=True)

#Lat = Y Long = X
```

```python
# Create a US
us = united_states.UnitedStates()
# Need a states list to check in 
states_list =  [i.name for i in us.states]
count = 0
def assign_states(lat,lon):
    global count
    returnVal = us.from_coords(lat,lon)
    count += 1
    if not count %500:
        print("Count at {}".format(count))
    if  not len(returnVal) > 0:
        return "NA"
    return returnVal[0].name
   
```

```python
%%time
w2['state'] = w2.apply(lambda row: assign_states(row.lat,row.lon),axis=1)
```

```python
w2.to_csv('data/weather_radar_stations_state.csv',index=False)
```
